***Because downloading and processing files with multiple categories took time, the above process was repeated for 1 selected category at a time and then all the files were combined to get final dataset.***

```python
import json
import os

def combine_jsonl_files(folder_path, output_file):
    with open(output_file, 'w') as output:
        for filename in os.listdir(folder_path):
            if filename.endswith(".jsonl"):
                file_path = os.path.join(folder_path, filename)
                with open(file_path, 'r') as infile:
                    for line in infile:
                        output.write(line)

if __name__ == "__main__":
    input_folder = "/content/drive/MyDrive/Thesis/Icecat/Category_datasets"
    output_file = "icecat.jsonl"

    combine_jsonl_files(input_folder, output_file)

print(f"Files combined successfully. Output written to {output_file}")
```

***Convert data in the required format***

```python
import json

def process_jsonl_file(input_file, output_file):
    categories_attributes = {}

    # Read the input JSONL file
    with open(input_file, 'r') as file:
        for line in file:
            data = json.loads(line.strip())

            # Get category and target_scores from the current record
            category = data.get('category', '')
            target_scores = data.get('target_scores', {})

            # If category is not in the dictionary, add it with the attributes
            if category not in categories_attributes:
                categories_attributes[category] = set(target_scores.keys())
            else:
                # Update the set of attributes for the category
                categories_attributes[category].update(target_scores.keys())

    # Read the input JSONL file again to update missing attributes
    with open(input_file, 'r') as file, open(output_file, 'w') as output:
        for line in file:
            data = json.loads(line.strip())

            # Get category and target_scores from the current record
            category = data.get('category', '')
            target_scores = data.get('target_scores', {})

            # Get the set of unique attributes for the category
            unique_attributes = categories_attributes.get(category, set())

            # Find missing attributes and append them with value 'n/a'
            missing_attributes = unique_attributes - set(target_scores.keys())
            for attribute in missing_attributes:
                target_scores[attribute] = {'n/a': 1}

            # Update the target_scores in the data
            data['target_scores'] = target_scores

            # Write the updated data to the output file
            output.write(json.dumps(data) + '\n')

# Example usage
process_jsonl_file('/content/Icecat_final.jsonl', 'output.jsonl')
```

***Convert all the attribute names and their target scores to lowecase***

```python
import json

def convert_values_to_lowercase(input_file, output_file):
    with open(input_file, 'r') as file, open(output_file, 'w') as output:
        for line in file:
            data = json.loads(line.strip())

            # Convert attribute values to lowercase
            for attribute, values in data['target_scores'].items():
                data['target_scores'][attribute] = {key.lower(): value for key, value in values.items()}

            # Write the updated data to the output file
            output.write(json.dumps(data) + '\n')

convert_values_to_lowercase('/content/final_icecat_complete.jsonl', 'output_lowercase.jsonl')
```

***Split file into train and test datasets***

```python
import json
import random

def load_jsonl_file(file_path):
    with open(file_path, 'r') as file:
        data = [json.loads(line) for line in file]
    return data

def save_jsonl_file(data, file_path):
    with open(file_path, 'w') as file:
        for line in data:
            file.write(json.dumps(line) + '\n')

def create_train_test_datasets(data):
    random.shuffle(data)
    train_examples = []
    test_examples = []
    categories = list(set([example['category'] for example in data]))
    for category in categories:
        category_examples = [example for example in data if example['category'] == category]
        train_examples.extend(category_examples[:int(len(category_examples) * 0.20)])
        # test_examples.extend(category_examples[int(len(category_examples) * 0.75):])

    return train_examples, test_examples

input_file_path = '/content/train.jsonl'
output_train_path = '/content/train_0.2.jsonl'
# output_test_path = '/content/test_dataset.jsonl'

data = load_jsonl_file(input_file_path)
train_dataset, test_dataset = create_train_test_datasets(data)

save_jsonl_file(train_dataset, output_train_path)
# save_jsonl_file(test_dataset, output_test_path)
```

***Finding statistics related to the datasets***

```python
import json

def find_unique_attributes_by_category(input_file):
    with open(input_file, 'r') as file:
        lines = file.readlines()

    unique_attributes_per_category = {}

    for line in lines:
        data = json.loads(line)
        category = data.get('category', '')
        target_scores = data.get('target_scores', {})
        attributes = set(target_scores.keys())

        # Update unique attributes for each category
        if category not in unique_attributes_per_category:
            unique_attributes_per_category[category] = attributes
        else:
            unique_attributes_per_category[category].update(attributes)

    return unique_attributes_per_category


input_file_path = '/content/output.jsonl'
# input_file_path = '/content/test12.jsonl'
result = find_unique_attributes_by_category(input_file_path)

# Print the unique attributes for each category
for category, attributes in result.items():
    print(f'Category: {category}, Unique Attributes: {len(attributes)}')
    print(attributes)
```

```python
import json

def find_unique_attribute_value_pairs(input_file):
    categories_attribute_values = {}

    # Read the input JSONL file
    with open(input_file, 'r') as file:
        for line in file:
            data = json.loads(line.strip())

            # Get category and target_scores from the current record
            category = data.get('category', '')
            target_scores = data.get('target_scores', {})

            # If category is not in the dictionary, add it with the attribute-value pairs
            if category not in categories_attribute_values:
                categories_attribute_values[category] = {attr: set(value.keys()) for attr, value in target_scores.items()}
            else:
                # Update the set of attribute-value pairs for the category
                for attr, value in target_scores.items():
                    categories_attribute_values[category].setdefault(attr, set()).update(value.keys())

    return categories_attribute_values

# Example usage
input_file = '/content/final_icecat_complete.jsonl'
unique_attribute_values = find_unique_attribute_value_pairs(input_file)

# Print the results
for category, attribute_values in unique_attribute_values.items():
    print(f"\nCategory: {category}")
    for attribute, values in attribute_values.items():
        print(f"{attribute}: {values}")
```
